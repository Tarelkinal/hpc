```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plotImage(image):
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(rgb_image)

def plotImages(image1, image2):
    plt.figure(figsize=(32, 32))
    plt.subplot(2, 2, 1)
    plotImage(image1)
    plt.subplot(2, 2, 2)
    plotImage(image2)
    plt.tight_layout()
    plt.show()
```

```python
image = cv2.imread("./cat_2.jpg")
np.shape(image)
```

# Используем медианный фильтр OpenCV

```python
filtered_image = cv2.medianBlur(image, 5)
```

```python
plotImages(image, filtered_image)
```
